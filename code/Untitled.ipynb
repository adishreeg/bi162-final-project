```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([175, 200, 200, 200, 200, 100, 150, 225, 225, 225, 250, 250, 250, 275, 275, 275,
    300, 300, 300, 300, 500, 500, 500, 400, 400, 400])
y = [0, 14, 4, 3, 7, 0, 0, 3, 10, 10, 14, 14, 8, 13, 15, 10, 7, 16, 8, 16, 22, 19, 22, 
     15, 18, 18]

a, b = np.polyfit(x, y, 1)
```

```python
plt.scatter(x, y, color='k', label='Recorded data')
plt.plot(x, a * x + b, color='b', label='Line of best fit')
plt.xlabel('Voltage (mV) of injected pulse')
plt.ylabel('Spikes / sec')
plt.legend()
plt.title('Firing rate response for T2 cell in leech ganglion')
plt.show()
```
